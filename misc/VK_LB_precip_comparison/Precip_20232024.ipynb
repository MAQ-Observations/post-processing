```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
vk = pd.read_csv('Veenkampen.csv',skiprows=2,header=None,delimiter=';')
vk['cum_sum'] = vk[1].cumsum()
vk[0] = pd.to_datetime(vk[0], format='%Y-%m-%d %H:%M')

lb = pd.read_csv('Loobos.csv',skiprows=2,header=None,delimiter=',')
print(lb)
lb['cum_sum'] = lb[1].cumsum()
lb[0] = pd.to_datetime(lb[0], format='%Y-%m-%d %H:%M')


plt.plot(vk[0],vk['cum_sum'],label='Veenkampen')
plt.plot(lb[0],lb['cum_sum'],label='Loobos')
plt.xlim([vk[0].values[0],vk[0].values[-1]])
plt.xticks(rotation = 25)
plt.ylabel('Cumulative precipitation [mm]')
plt.xlabel('Date')
plt.legend(loc='best')
plt.grid()
plt.show()

print('Cumulative precipitation Veenkampen = '+str(np.round(vk['cum_sum'].astype(float).values[-1],0))+' mm')
print('Cumulative precipitation Loobos = '+str(np.round(lb['cum_sum'].astype(float).values[-1],0))+' mm')
```
